```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from datascience import *
```

```python
aquality_outside = pd.read_csv('aquality outside.csv').iloc[30:330]
aquality_inside = pd.read_csv('aquality.csv')
```

```python
#temperature

plt.plot(aquality_inside.index, aquality_inside['temperature'], label='Inside')
plt.plot(aquality_outside.index-30, aquality_outside['temperature'], label='Outside')

plt.xlabel('Time')
plt.ylabel('Temperature')
plt.title('Inside vs Outside Temperature')
plt.legend()
plt.show()
```

```python
#gas

plt.plot(aquality_inside.index, aquality_inside['gas'], label='Inside')
plt.plot(aquality_outside.index-30, aquality_outside['gas'], label='Outside')

plt.xlabel('Time')
plt.ylabel('gas')
plt.title('Inside vs Outside gas')
plt.legend()
plt.show()
```

```python
#humidity

plt.plot(aquality_inside.index, aquality_inside['humidity'], label='Inside')
plt.plot(aquality_outside.index-30, aquality_outside['humidity'], label='Outside')

plt.xlabel('Time')
plt.ylabel('humidity')
plt.title('Inside vs Outside humidity')
plt.legend()
plt.show()
```

```python
#pressure

plt.plot(aquality_inside.index, aquality_inside['pressure'], label='Inside')
plt.plot(aquality_outside.index-30, aquality_outside['pressure'], label='Outside')

plt.xlabel('Time')
plt.ylabel('pressure')
plt.title('Inside vs Outside pressure')
plt.legend()
plt.show()
```

```python
#altitude

plt.plot(aquality_inside.index, aquality_inside['altitude'], label='Inside')
plt.plot(aquality_outside.index-30, aquality_outside['altitude'], label='Outside')

plt.xlabel('Time')
plt.ylabel('altitude')
plt.title('Inside vs Outside altitude')
plt.legend()
plt.show()
```

```python
#histogram for temperature

plt.hist([aquality_outside['temperature'], aquality_inside['temperature']], label=['Outside', 'Inside'])
plt.xlabel('temperature')
plt.ylabel('Frequency')
plt.title('Inside vs Outside temperature')
plt.legend()
plt.show()
```

```python
#histogram for gas

plt.hist([aquality_outside['gas'], aquality_inside['gas']], label=['Outside', 'Inside'])
plt.xlabel('gas')
plt.ylabel('Frequency')
plt.title('Inside vs Outside gas')
plt.legend()
plt.show()
```

```python
print('outside mean:', np.mean(aquality_outside['gas']))
print('inside mean:', np.mean(aquality_inside['gas']))
print('mean diff:', np.mean(aquality_outside['gas'])-np.mean(aquality_inside['gas']))
print('')

stdev_outside = np.std(aquality_outside['gas']) / np.sqrt(len(aquality_outside['gas']))
stdev_inside = np.std(aquality_inside['gas']) / np.sqrt(len(aquality_inside['gas']))
print('stdev_outside:', stdev_outside)
print('stdev_inside:', stdev_inside)
print('')

mean_diff = np.mean(aquality_outside['gas'])-np.mean(aquality_inside['gas'])
print(mean_diff / stdev_outside)
```
