```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from datascience import *
%matplotlib inline
```

```python
aquality_outside = pd.read_csv('/aquality_outside.csv').iloc[31:330]
aquality_inside = pd.read_csv('/aquality.csv').iloc[1:300]
```

```python
outside_2 = pd.read_csv('/outside.csv').iloc[31:330]
inside_2 = pd.read_csv('/inside.csv').iloc[1:300]
```

```python
np.array(aquality_inside.index)
```

```python
#temperature

plt.plot(np.array(aquality_inside.index)-30, aquality_inside['temperature'], label='Inside1')
plt.plot(np.array(inside_2.index)-30, inside_2['Temp (C)'], label='Inside2')

plt.xlabel('Time')
plt.ylabel('Temperature')
plt.title('Inside1 vs Inside2 Temperature')
plt.legend()
plt.show()
```

```python
#gas

plt.plot(np.array(aquality_inside.index)-30, aquality_inside['gas'], label='Inside1')
plt.plot(np.array(inside_2.index)-30, inside_2['Gas (Ohm)'], label='Inside2')

plt.xlabel('Time')
plt.ylabel('gas')
plt.title('Inside1 vs Inside2 gas')
plt.legend()
plt.show()
```

```python
#humidity

plt.plot(np.array(aquality_inside.index)-30, aquality_inside['humidity'], label='Inside1')
plt.plot(np.array(inside_2.index)-30, inside_2['Humidity (%)'], label='Inside2')

plt.xlabel('Time')
plt.ylabel('humidity')
plt.title('Inside1 vs Inside2 humidity')
plt.legend()
plt.show()
```

```python

#histogram for temperature

plt.hist([inside_2['Temp (C)'], aquality_inside['temperature']], label=['Inside2', 'Inside1'])
plt.xlabel('temperature')
plt.ylabel('Frequency')
plt.title('Inside1 vs Inside2 temperature')
plt.legend()
plt.show()
```

```python

#histogram for gas

plt.hist([inside_2['Gas (Ohm)'], aquality_inside['gas']], label=['Inside2', 'Inside1'])
plt.xlabel('gas')
plt.ylabel('Frequency')
plt.title('Inside1 vs Inside2 gas')
plt.legend()
plt.show()
```

```python
print('inside2 mean:', np.mean(inside_2['Gas (Ohm)']))
print('inside1 mean:', np.mean(aquality_inside['gas']))
print('mean diff:', np.mean(inside_2['Gas (Ohm)'])-np.mean(aquality_inside['gas']))
print('')

stdev_inside2 = np.std(inside_2['Gas (Ohm)']) / np.sqrt(len(inside_2['Gas (Ohm)']))
stdev_inside1 = np.std(aquality_inside['gas']) / np.sqrt(len(aquality_inside['gas']))
print('stdev_inside2:', stdev_inside2)
print('stdev_inside1:', stdev_inside1)
print('')

mean_diff = np.mean(inside_2['Gas (Ohm)'])-np.mean(aquality_inside['gas'])
print(mean_diff / stdev_inside2)
```

```python
metal_aquality = pd.read_csv('/metal_aquality.csv').iloc[60:360]
```

```python
#temperature and humidity

plt.plot(metal_aquality['temperature'], metal_aquality['humidity'], label='corrolation')


plt.xlabel('temperature')
plt.ylabel('humidity')
plt.title('temperature vs humidity metal shop')
plt.legend()
plt.show()
```

```python
#doest look linear
#hidden variable is time
```

```python
#this plot shows temperature vs humidity, they are corrolated in the sense that their information were both retrieved by the same sensor at the same point in time
# its not linear because it doesnt pas the vertical line test
# I would implement these in code like this:
import numpy as np
from scipy.stats import pearsonr

temperature = metal_aquality['temperature']
humidity = metal_aquality['humidity']

# Calculating Pearson's correlation
correlation_coefficient, p_value = pearsonr(temperature, humidity)

print(f"Pearson's correlation coefficient: {correlation_coefficient}")
print(f"P-value: {p_value}")
```
